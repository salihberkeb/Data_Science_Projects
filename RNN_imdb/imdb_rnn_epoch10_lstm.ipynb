```python
from tensorflow.keras.layers import SimpleRNN,Embedding,Dense,Bidirectional,LSTM,Dropout
from tensorflow.keras.models import Sequential
from keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
```

```python
num_features=1000
maxlen=500
```

```python

(x_train,y_train),(x_test,y_test)=imdb.load_data(num_words=1000,maxlen=500)
```

```python
batch_size=32
```

```python
x_train=sequence.pad_sequences(x_train,maxlen=maxlen)
x_test=sequence.pad_sequences(x_test,maxlen=maxlen)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
model=Sequential()

model.add(Embedding(num_features,32,input_length=maxlen))
model.add(Bidirectional(LSTM(128,activation="tanh")))
model.add(Dropout(0.2))
model.add(Dense(1,activation="tanh"))

model.summary()
```

```python
model.compile(optimizer="rmsprop",
              loss="binary_crossentropy",
              metrics=["accuracy"])

history=model.fit(x_train,y_train,
          epochs=10,
          batch_size=128,
          validation_data=(x_test,y_test),
          validation_split=0.2)

```

```python
acc=history.history["accuracy"]
val_acc=history.history["val_accuracy"]
loss=history.history["loss"]
val_loss=history.history["val_loss"]
```
