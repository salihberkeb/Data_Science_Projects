```python
raw_data=pd.get_dummies(new_Dataframe,drop_first=0)
```

```python
y=raw_data["Price"].values
x=raw_data.drop("Price",axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=365)
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
scaler=MinMaxScaler()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
raw_data["Price"]=raw_data["Price"].astype(np.int64)
```

```python
raw_data.info()
```

```python
raw_data=raw_data.astype(np.int64)
```

```python
raw_data.drop("Price",axis=1).info()
```

```python
raw_data
```

```python
x_train.shape
```

```python

model=Sequential()

model.add(Dense(78,activation="relu"))
model.add(Dense(39,activation="relu"))
model.add(Dense(39,activation="relu"))
model.add(Dense(15,activation="relu"))
model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")

```

```python
earlyStopping=EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=200)
```

```python
model.fit(x=x_train,y=y_train,validation_data=(x_test,y_test),epochs=400,verbose=1)
```

```python
kayiprveri=pd.DataFrame(model.history.history)
```
